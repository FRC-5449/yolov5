```python
import requests
import time
start_time = time.time()
t = []
while True:
    resp = requests.get("http://127.0.0.1:8880/getResult")
    if resp.content:
        print(resp.content)
    t.append(time.time()-start_time)
    start_time = time.time()
```

```python
print("Average request round time", sum(t)/len(t))
```
